# Análisis Univariado

Las primeras visualizaciones nos puede servir para ver que hay bastante variedad en cuanto a las longitudes del sépalo de cada especie. Podemos ver que hay mucha diferencia en cuanto a longitud de las distintas especies, pero tambien hay una importante diferencia entre las medidas en cm de cada sépalo. 



No podemos decir lo mismo en cuanto al ancho del sépalo de cada especie; ya que como vemos, el ancho es algo similar, aún cuando la especie no es la misma. Esto nos da la pauta de que este dato, no es muy útil a la hora de hacer algun modelo de predicción.



# Análisis Multivariado

### Hagamos un análisis con mas de dos variantes, como por ejemplo, la longitud y la especie, y tambien el ancho y la espcie. Esto no dará una visión mas específica.

#### Longitud por especie

Estos gráficos nos muestran la difenrecia de la longitud del sépalo de cada especie. Podemos ver que hay una importante difenrecia entre cada una de las especies. Vemos que la especie "Setosa" es la que tiene la longitud de menor tamaño, mientras que la especie "Virginica" es la de mayo tamaño.



#### Ancho por especie

Ahora el mismo análisis pero con el ancho de cada especie. Tambien nos deja una claridad de que depende la especie, el ancho varía también.



# Ahora hgamos el análisis del gráfico multivariado

Bien, estos gráficos podemos decir que "van de la mano" porque vemos en el primer gráfico lo que el mapa del calor nos muestra en números. Mediante la correlación lineal de las variables podemos ver los resultados en números, como hay algunos casi llegando al entéro, mientras que hay otros en negativo (-). Esto nos muestra una correlación inversa, y un buen ejemplo para entender es: Un niño cuanto mas juega playStations, su rendimiento escolar disminuye. Mientras que las otras correlaciones, no negativas, son lineales directas.
Pero aclaramos que cuando el número sea negativo, eso no quiere decir que no haya relación.

Y podemos concluir que hay algunos casos para especies que, cuanto mas aumenta su longitud, mas aumenta su ancho (correlación lineal directa), o visceversa; mientras que por el contrario, en otros casos, cuando mas aumenta su longitud, mas disminuye su ancho (correlación inversa), o visceversa.
